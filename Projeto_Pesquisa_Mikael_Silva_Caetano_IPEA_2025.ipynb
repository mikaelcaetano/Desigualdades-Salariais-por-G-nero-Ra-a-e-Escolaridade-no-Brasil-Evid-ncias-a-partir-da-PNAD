Desigualdades Salariais por Gênero, Raça e Escolaridade no Brasil: Evidências a partir da PNAD Contínua 2025/2

## 1. Pergunta de Pesquisa
O objetivo deste estudo é investigar as desigualdades salariais no Brasil, considerando as variáveis de gênero, cor/raça e escolaridade. A principal questão é: quais são as diferenças salariais entre os grupos e em que medida a interseccionalidade entre gênero, raça e escolaridade intensifica ou atenua essas desigualdades no rendimento do trabalho?

## 2. Metodologia e Base de Dados
### 2.1. Base de Dados
Este estudo utiliza a base de dados da PNAD Contínua Trimestral, referente ao 2º trimestre de 2025, acessada por meio da biblioteca pnadium. As variáveis selecionadas incluem:

VD4008: Renda habitual do trabalho principal

VD4002: Condição de ocupação

V2007: Sexo

V2010: Cor/raça

V3003A: Escolaridade

V2009: Idade

A análise é restrita a indivíduos entre 18 e 64 anos, que estão ocupados e possuem renda positiva.

### 2.2. Método
A abordagem metodológica consiste nas seguintes etapas:

Estatísticas Descritivas: Análise da renda média segregada por:

Sexo

Raça

Escolaridade

Combinações interseccionais de sexo, raça e escolaridade (sexo × raça × escolaridade).

Visualizações Gráficas: Geração de gráficos para ilustrar as desigualdades salariais:

Gráficos de barras para comparar a renda média por grupo.

Heatmap (mapa de calor) para observar a interação entre escolaridade, sexo e raça.

Boxplots e violin plots para ilustrar a distribuição de renda por grupo.

Indicadores de Desigualdade:

Índice de Gini por sexo, raça e escolaridade, para medir a desigualdade salarial.

Prêmio de Escolaridade: variação percentual na renda média conforme o nível de escolaridade.

Diferença percentual no prêmio de escolaridade entre grupos.

### 2.3. Tratamento das Variáveis
O tratamento dos dados envolve:

Remoção de rendas inválidas (≤ 0).

Filtro por idade, restringindo a faixa etária entre 18 e 64 anos, para focar na população economicamente ativa.

Recodificação das variáveis (sexo, raça, escolaridade) para categorias mais intuitivas.

Agrupamento em variáveis compostas para representar interações entre sexo e raça (por exemplo, "Homem Branco", "Mulher Negra").

## 3. Análise dos Resultados
### 3.1. Resultados Iniciais
As análises preliminares revelam desigualdades salariais sistemáticas:

Mulheres ganham, em média, menos do que os homens.

Pessoas pretas e pardas apresentam rendimentos inferiores aos de pessoas brancas, em todos os níveis de escolaridade.

O rendimento cresce substancialmente com a escolaridade, mas as desigualdades de gênero e raça persistem em todos os níveis de educação.

### 3.2. Interseccionalidade
Ao considerar a interseccionalidade entre sexo, raça e escolaridade, observa-se que:

Mulheres negras e pardas têm a menor renda média, mesmo com níveis educacionais semelhantes aos de outros grupos.

Homens brancos ocupam, de maneira consistente, o topo da distribuição salarial, independentemente do nível de escolaridade.

Esses achados indicam que a combinação de gênero e raça agrava as desigualdades salariais.

### 3.3. Escolaridade como Redutor Parcial da Desigualdade
Embora a escolaridade seja um fator positivo para a renda de todos os grupos, ela não elimina as desigualdades estruturais. A escolaridade atua como um redutor parcial da desigualdade, mas o diferencial racial e de gênero se mantém ao longo dos níveis educacionais. Além disso:

Mulheres e pessoas negras apresentam um menor prêmio de escolaridade, ou seja, a variação da renda com o aumento da escolaridade é menor para esses grupos.

A concentração de renda continua mais alta nos grupos privilegiados, especialmente entre os homens brancos.

### 3.4. Desigualdade Interna
O Índice de Gini revela a intensidade das desigualdades dentro de cada grupo:

A desigualdade é maior entre os homens, refletindo uma maior dispersão salarial dentro desse grupo.

A desigualdade é mais pronunciada entre os brancos, indicando uma dispersão salarial elevada dentro dos grupos mais privilegiados.

Por outro lado, a desigualdade interna é menor nos grupos com menos escolaridade, sugerindo uma distribuição de renda mais homogênea nas camadas mais baixas.

### 3.5. Conclusão
Os resultados confirmam que:

Gênero, raça e escolaridade são determinantes fundamentais das desigualdades salariais no Brasil.

A interseccionalidade entre esses fatores agrava as disparidades, com mulheres negras e pardas sendo os grupos mais penalizados.

A escolaridade contribui para a redução das desigualdades, mas não é suficiente para eliminar as disparidades estruturais, principalmente em relação ao gênero e raça.

Esses achados sugerem que políticas públicas voltadas para a igualdade salarial devem considerar a interseccionalidade e trabalhar tanto na ampliação do acesso à educação quanto na promoção da equidade de gênero e racial no mercado de trabalho

## 4. Código Phyton
### 4.1. Comentários sobre o código:
Instalação e Importação de Bibliotecas: O código começa com a instalação da biblioteca pnadium e a importação de outras bibliotecas úteis para análise e visualização de dados, como pandas, numpy, seaborn, matplotlib e pnadium.

Download e Seleção de Dados: A função pnadium.trimestral.download() é usada para baixar os dados da PNAD com o ano e trimestre especificado, e apenas as colunas necessárias são selecionadas.

Limpeza dos Dados: São aplicados filtros para considerar apenas os trabalhadores ocupados, com renda válida e dentro da faixa etária de 18 a 64 anos.

Recode de Variáveis: As variáveis sexo, raca e escolaridade são recodificadas para representações mais compreensíveis, usando mapeamento de valores numéricos para categorias.

Estatísticas Descritivas: São calculadas as médias salariais por sexo, raça, escolaridade e grupo (sexo × raça), e também o tamanho da amostra final.

Visualizações: São gerados gráficos de barras, heatmaps e boxplots para visualizar as distribuições e médias salariais por diferentes variáveis.

Índice de Gini: O índice de Gini é calculado para medir a desigualdade salarial por sexo, raça e escolaridade.

Prêmio de Escolaridade: Calcula-se o prêmio de escolaridade, que representa a variação percentual da renda conforme o nível de escolaridade. Gráficos são gerados para mostrar essa variação.

```python
# ============================================================
# PROJETO – Desigualdades salariais por gênero e raça
# ============================================================

# Instalando a biblioteca pnadium para acessar a base de dados da PNAD
!pip install pnadium

# Importando as bibliotecas necessárias para análise
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pnadium

# ------------------------------------------------------------
# 1. DOWNLOAD E SELEÇÃO DA BASE
# ------------------------------------------------------------

# Baixando os dados da PNAD do segundo trimestre de 2025 e selecionando as colunas relevantes
pnad = pnadium.trimestral.download(
    2025,
    2,
    colunas=["VD4008", "VD4002", "V2007", "V2010", "V2009", "V3003A"]
)

# ------------------------------------------------------------
# 2. LIMPEZA BÁSICA DOS DADOS
# ------------------------------------------------------------

# Criando uma cópia do DataFrame original para evitar alterar os dados originais
df = pnad.copy()

# Filtrando os dados para considerar apenas os trabalhadores "ocupados" (VD4002 == 1)
df = df[df["VD4002"] == 1]

# Filtrando para considerar apenas rendas válidas (renda maior que zero)
df = df[df["VD4008"] > 0]

# Filtrando para considerar apenas idades entre 18 e 64 anos
df["V2009"] = pd.to_numeric(df["V2009"], errors="coerce")  # Convertendo para numérico, com erro tratado
df = df[(df["V2009"] >= 18) & (df["V2009"] <= 64)]

# ------------------------------------------------------------
# 3. RECODIFICAÇÃO DAS VARIÁVEIS
# ------------------------------------------------------------

# Recode da variável 'sexo' (1 = Homem, 2 = Mulher)
df["sexo"] = df["V2007"].map({1: "Homem", 2: "Mulher"})

# Recode da variável 'raça' (1 = Branco, 2 = Preto, 3 = Pardo, 4 = Amarelo, 5 = Indígena, 9 = Ignorado)
df["raca"] = df["V2010"].map({
    1: "Branco",
    2: "Preto",
    3: "Pardo",
    4: "Amarelo",
    5: "Indígena",
    9: "Ignorado"
})

# Recode da variável 'escolaridade' (categorias de 2 a 11 correspondem a diferentes níveis de escolaridade)
escola_map = {
    2: "Pré-escola",
    3: "Alfabetização de adultos",
    4: "Ensino fundamental",
    5: "EJA fundamental",
    6: "Ensino médio",
    7: "EJA médio",
    8: "Superior - graduação",
    9: "Especialização",
    10: "Mestrado",
    11: "Doutorado"
}
df["escolaridade"] = df["V3003A"].map(escola_map)

# Criando uma nova coluna para o grupo, que combina sexo e raça
df["grupo"] = df["sexo"] + " - " + df["raca"]

# Definindo a variável 'renda' (VD4008)
df["renda"] = df["VD4008"]

# Removendo linhas com dados faltantes em 'sexo', 'raca' ou 'escolaridade'
df = df.dropna(subset=["sexo", "raca", "escolaridade"])

# ------------------------------------------------------------
# 4. ESTATÍSTICAS DESCRITIVAS
# ------------------------------------------------------------

# Exibindo a média salarial por sexo
print("\n=== Média salarial por sexo ===")
print(df.groupby("sexo")["renda"].mean())

# Exibindo a média salarial por raça
print("\n=== Média salarial por raça ===")
print(df.groupby("raca")["renda"].mean())

# Exibindo a média salarial por escolaridade
print("\n=== Média salarial por escolaridade ===")
print(df.groupby("escolaridade")["renda"].mean())

# Exibindo a média salarial por grupo (sexo × raça)
print("\n=== Média salarial por grupo (sexo × raça) ===")
print(df.groupby("grupo")["renda"].mean().sort_values())

# Exibindo o tamanho da amostra final após os filtros
print("\nTamanho da amostra final:", len(df))

# ------------------------------------------------------------
# 5. VISUALIZAÇÕES
# ------------------------------------------------------------

# Gráfico de barras com a média salarial por grupo (sexo × raça)
plt.figure(figsize=(10,5))
df.groupby("grupo")["renda"].mean().sort_values().plot(kind="bar", color='skyblue')
plt.title("Média Salarial por Sexo e Raça – PNAD 2025/2")
plt.ylabel("Renda (R$)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Heatmap com a média salarial por escolaridade, sexo e raça
tabela = df.pivot_table(
    values="renda",
    index="escolaridade",
    columns=["sexo", "raca"],
    aggfunc="mean"
)

plt.figure(figsize=(14, 8))
sns.heatmap(tabela, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Renda Média por Escolaridade, Sexo e Raça")
plt.tight_layout()
plt.show()

# Boxplot para a distribuição da renda por sexo
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x="sexo", y="renda", palette="Set2")
plt.title("Distribuição da Renda por Sexo")
plt.show()

# Violinplot para a distribuição da renda por raça
plt.figure(figsize=(10,5))
sns.violinplot(data=df, x="raca", y="renda", palette="Set3")
plt.title("Distribuição da Renda por Raça")
plt.show()

# ------------------------------------------------------------
# 6. CÁLCULO DO ÍNDICE DE GINI
# ------------------------------------------------------------

# Função para calcular o índice de Gini
def gini(arr):
    arr = arr[arr >= 0]  # Excluindo valores negativos
    if len(arr) == 0:
        return np.nan
    arr = np.sort(arr)  # Ordenando os valores
    n = len(arr)
    cum = np.cumsum(arr)  # Cálculo da soma cumulativa
    return (n + 1 - 2 * np.sum(cum / cum[-1])) / n  # Fórmula do Gini

# Calculando o índice de Gini por sexo, raça e escolaridade
gini_sexo = df.groupby("sexo")["renda"].apply(gini)
gini_raca = df.groupby("raca")["renda"].apply(gini)
gini_escolaridade = df.groupby("escolaridade")["renda"].apply(gini)

# Exibindo os resultados do índice de Gini
print("\nGINI por Sexo:\n", gini_sexo)
print("\nGINI por Raça:\n", gini_raca)
print("\nGINI por Escolaridade:\n", gini_escolaridade)

# ------------------------------------------------------------
# 7. CÁLCULO DO PRÊMIO DA ESCOLARIDADE
# ------------------------------------------------------------

# Calculando a variação percentual do prêmio de escolaridade
media_esc = df.groupby("escolaridade")["renda"].mean()
premio_esc = media_esc.pct_change() * 100  # Variação percentual

# Exibindo o prêmio de escolaridade
print("\nPrêmio da Escolaridade (%):")
print(premio_esc.round(2))

# Prêmio de escolaridade por raça
premio_raca = df.groupby(["raca", "escolaridade"])["renda"].mean().groupby(level=0).pct_change() * 100
print("\nPrêmio por raça (%):")
print(premio_raca.round(2))

# Prêmio de escolaridade por sexo
premio_sexo = df.groupby(["sexo", "escolaridade"])["renda"].mean().groupby(level=0).pct_change() * 100
print("\nPrêmio por sexo (%):")
print(premio_sexo.round(2))

# ------------------------------------------------------------
# 8. GRÁFICOS PARA GINI E PRÊMIO DE ESCOLARIDADE
# ------------------------------------------------------------

# 8.1 Gráfico do Índice de Gini por Sexo, Raça e Escolaridade
# Índice de Gini por Sexo
plt.figure(figsize=(10,5))
gini_sexo.plot(kind="bar", color=['skyblue', 'orange'])
plt.title("Índice de Gini por Sexo")
plt.ylabel("Índice de Gini")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Índice de Gini por Raça
plt.figure(figsize=(10,5))
gini_raca.plot(kind="bar", color=['lightgreen', 'lightcoral', 'lightgoldenrodyellow', 'lightpink', 'lightblue'])
plt.title("Índice de Gini por Raça")
plt.ylabel("Índice de Gini")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Índice de Gini por Escolaridade
plt.figure(figsize=(10,5))
gini_escolaridade.plot(kind="bar", color='mediumpurple')
plt.title("Índice de Gini por Escolaridade")
plt.ylabel("Índice de Gini")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# 8.2 Gráfico do Prêmio de Escolaridade
# Prêmio de Escolaridade
plt.figure(figsize=(12,6))
premio_esc.plot(kind="bar", color='steelblue')
plt.title("Variação do Prêmio de Escolaridade (%)")
plt.ylabel("Prêmio (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Prêmio de Escolaridade por Raça
plt.figure(figsize=(12,6))
premio_raca.unstack().plot(kind="bar", stacked=False, figsize=(14,7))
plt.title("Variação do Prêmio de Escolaridade por Raça (%)")
plt.ylabel("Prêmio (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Prêmio de Escolaridade por Sexo
plt.figure(figsize=(12,6))
premio_sexo.unstack().plot(kind="bar", stacked=False, figsize=(14,7))
plt.title("Variação do Prêmio de Escolaridade por Sexo (%)")
plt.ylabel("Prêmio (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
